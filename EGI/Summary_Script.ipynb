```python
import openai
import wget
import pathlib
import pdfplumber
import numpy as np

def getPaper(paper_url, filename="random_paper.pdf"):
    """
    Downloads a paper from it's arxiv page and returns
    the local path to that file.
    """
    downloadedPaper = wget.download(paper_url, filename)    
    downloadedPaperFilePath = pathlib.Path(downloadedPaper)

    return downloadedPaperFilePath


def showPaperSummary(paperContent):
    tldr_tag = "\n tl;dr:"
    openai.organization = 'org-HOZhp7IqkIkQxmCY6glZLbIK'
    openai.api_key = "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"
    engine_list = openai.Engine.list() 
    
    for page in paperContent:    
        text = page.extract_text() + tldr_tag
        response = openai.Completion.create(engine="text-davinci-002",prompt=text,temperature=0.6,
            max_tokens=128,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            stop=["\n"]
        )
        with open("Fashion-interviste.md", "a") as text_file:
            text_file.write(response["choices"][0]["text"])
            text_file.write("\n")


paperContent = pdfplumber.open("/home/eduard/Github/University/EGI/Consegna 2/Fashion-interviste.pdf").pages
showPaperSummary(paperContent)
```
