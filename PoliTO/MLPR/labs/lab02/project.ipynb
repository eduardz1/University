```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "data/trainData.txt"

dataset = np.loadtxt(PATH, delimiter=",")

# splits the dataset into features and labels
X = dataset[:, :-1]
y = dataset[:, -1]

print(X.shape, y.shape)
print(X[:5], y[:5])
```

```python
# Plots histograms of the features

for i in range(X.shape[1]):
    plt.subplot(2, 3, i + 1)
    plt.hist(X.T[:, y == 0][i], bins=10, density=True, alpha=0.4, label="Fake")
    plt.hist(X.T[:, y == 1][i], bins=10, density=True, alpha=0.4, label="Genuine")
    plt.xlabel(f"Feature {i + 1}")
    plt.legend()
    plt.tight_layout()

plt.show()
```

```python
# Plots of pairwise scatter plots of the features

for i in range(X.shape[1]):
    for j in range(i + 1, X.shape[1]):
        if i == j:
            continue
        plt.scatter(X[y == 0][:, i], X[y == 0][:, j], alpha=0.4, label="Fake")
        plt.scatter(X[y == 1][:, i], X[y == 1][:, j], alpha=0.4, label="Genuine")
        plt.xlabel(f"Feature {i + 1}")
        plt.ylabel(f"Feature {j + 1}")
        plt.legend()
        plt.tight_layout()
        plt.show()
```

# Histograms

We can observe that the classes overlap completely for features 1 and 2, partially for features 3 and 4, while features 5 and 6 have better separation

# Scatter plots

There is no good separation between the first and second feature while all the others offer better separation but not always one that can be solved with a linear classifier

## Third and Fourth features

The classes overlap slightly in the `y = x` region, they seem to have similar mean and variance with two distinct clusters

## Fifth and Sixth features

The classes overlap only lightly in the border regions of the clusters, we can recognize 4 distinct clusters for the `Fake` and 4 different ones for the `Genuine` labels
